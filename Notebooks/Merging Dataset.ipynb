## Merging of Datasets


```python
import pandas as pd
from pymongo import MongoClient
```

```python
csr_df = pd.read_csv('preprocessed_content.csv')
print("CSR DataFrame loaded with shape:", csr_df.shape)
```

```python
client = MongoClient("mongodb://localhost:27017/")
db = client["FinanceDB"]
collection = db["SP500_Companies"]
```

```python
reddit_df = pd.DataFrame(list(collection.find()))
print("Reddit DataFrame loaded with shape:", reddit_df.shape)
```

```python
reddit_df.rename(columns={'company': 'Company'}, inplace=True)
```

```python
reddit_summary = reddit_df.groupby('Company').agg(
    avg_sentiment=('sentiment', 'mean'),
    post_count=('sentiment', 'count')
).reset_index()

print(reddit_summary.head())
```
